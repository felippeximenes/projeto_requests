# Paginando repositorios

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'

url
```

```python
repos_list = []
for page_num in range(1, 7):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list
```

```python
len(repos_list)
```

```python
len(repos_list[0])
```

### Nomes dos repositórios

```python
repos_list
```

```python
repos_list[0][2]['name']
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
repos_name[:10]
```

```python
len(repos_name)
```

# Linguagens dos repositorios

```python
repos_list[1][1]['language']
```

```python
repos_language = []
for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
```

```python
len(repos_language)
```

```python
repos_language
```

# Criando um DataFrame

```python
import pandas as pd
```

```python
dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name
dados_amz['language'] = repos_language
```

```python
dados_amz
```

### Salvando o DataFrame

```python
dados_amz.to_csv('amazon.csv')
```
